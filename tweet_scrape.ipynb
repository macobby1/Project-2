```python
csv_file = "counts.csv"
csv_columns = ['word', 'count']
try:
    with open(csv_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in counts:
            writer.writerow(data)
except IOError:
    print("I/O error")
```
